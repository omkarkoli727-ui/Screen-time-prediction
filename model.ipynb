```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression 
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import r2_score,mean_squared_error 
from sklearn.preprocessing import StandardScaler
```

```python
data=pd.read_csv(r"final_data.csv")
```

```python
X=data.drop(["Stress_Level"],axis=1)
```

```python
X.shape
```

```python
y=data["Stress_Level"]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
model_LR=LinearRegression().fit(X_train,y_train)
y_pred_LR=model_LR.predict(X_test)
LinearRegression_r2=r2_score(y_test,y_pred_LR) 
print(LinearRegression_r2)
```

```python
from sklearn.svm import SVR
model_SVR=SVR().fit(X_train,y_train)
y_pred_SVR=model_SVR.predict(X_test)
SVR_r2=r2_score(y_test,y_pred_SVR) 
print(SVR_r2)
```

```python
from sklearn.ensemble import RandomForestRegressor
model_RFR=RandomForestRegressor().fit(X_train,y_train)
y_pred_RFR=model_RFR.predict(X_test)
model_RFR_r2=r2_score(y_test,y_pred_RFR) 
print(model_RFR_r2)
```

```python
from sklearn.tree import DecisionTreeRegressor
model_DTR=DecisionTreeRegressor().fit(X_train,y_train)
y_pred_DTR=model_DTR.predict(X_test)
model_DTR_r2=r2_score(y_test,y_pred_DTR) 
print(model_DTR_r2)
```
