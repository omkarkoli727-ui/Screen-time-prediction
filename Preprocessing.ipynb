```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
data=pd.read_csv("Smartphone_Usage_Productivity_Dataset_50000.csv")
```

```python
data.info()
```

Selecting the Object column to observe the column is important for prediction or not and identify the  Encoding type for the column.

```python
data["User_ID"].describe()
```

```python
data=data.drop(["User_ID"],axis=1)
```

There is no use of User Id column it's better to drop it

```python
data["Occupation"].unique()
```

```python
data["Occupation"].value_counts() #One hot encoding
```

```python
data=pd.get_dummies(data,columns=["Occupation"],dtype=int)
```

```python
data["Gender"].unique() #label Enconding
```

```python
data["Gender"]=LabelEncoder().fit_transform(data["Gender"])
```

```python
data["Device_Type"].unique()
```

```python
data["Device_Type"].value_counts() #Label Enconding
```

```python
data["Device_Type"]=LabelEncoder().fit_transform(data["Device_Type"])
```

```python
data=data.drop_duplicates()
```

```python
data.info()
```

```python
data.to_csv("final_data.csv", index=False)
```
